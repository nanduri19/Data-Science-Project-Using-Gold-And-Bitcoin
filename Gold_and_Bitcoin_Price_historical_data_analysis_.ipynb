## *Gold historical data*

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# importing package such as pandas, matplot, etc.,
import pandas as pd_Price_Of_Historicaldata
import numpy as mu_Price_Of_Historicaldata
import plotly.graph_objects as gom_Price_Of_Historicaldata
import matplotlib.pyplot as pm_Price_Of_Historicaldata
import plotly.express as ex_Price_Of_Historicaldata

GOld_Price_Of_Historicaldata = pd_Price_Of_Historicaldata.read_csv("/content/drive/MyDrive/Gold Futures Historical Data UK.csv")
print("Shape details: ",GOld_Price_Of_Historicaldata.shape)
print("Columns details: ",GOld_Price_Of_Historicaldata.columns)
```

```python
#### null
GOld_Price_Of_Historicaldata.replace([mu_Price_Of_Historicaldata.inf, -mu_Price_Of_Historicaldata.inf], mu_Price_Of_Historicaldata.nan, inplace=True)
print(GOld_Price_Of_Historicaldata.isna().any(axis=1).sum(), "count of null rows")
```

```python
GOld_Price_Of_Historicaldata.dropna(inplace=True)
GOld_Price_Of_Historicaldata.info()
```

```python
GOld_Price_Of_Historicaldata.duplicated().any()
```

```python
GOld_Price_Of_Historicaldata
```

```python

FI= gom_Price_Of_Historicaldata.Figure([gom_Price_Of_Historicaldata.Scatter(x=GOld_Price_Of_Historicaldata['Date'], y=GOld_Price_Of_Historicaldata['Price'])])
FI.update_layout(title={'text': "Gold price ranges Year wise"})
FI.show()
```

```python

FI= gom_Price_Of_Historicaldata.Figure([gom_Price_Of_Historicaldata.Scatter(x=GOld_Price_Of_Historicaldata['Date'], y=GOld_Price_Of_Historicaldata['Change %'])])
FI.update_layout(title={'text': "Gold price changes in percentage"})
FI.show()
```

```python
# Candlestick chart for gold stock data

F_gom_Price_Of_Historicaldata = gom_Price_Of_Historicaldata.Figure(data=[gom_Price_Of_Historicaldata.Candlestick(x=GOld_Price_Of_Historicaldata['Date'],
                open=GOld_Price_Of_Historicaldata['Open'],
                high=GOld_Price_Of_Historicaldata['High'],
                low=GOld_Price_Of_Historicaldata['Low'],
                close=GOld_Price_Of_Historicaldata['Price'])])

F_gom_Price_Of_Historicaldata.update_layout(title='Gold Stock Price Candlestick Chart',
                  xaxis_title='Date',
                  yaxis_title='Gold Stock Price')

F_gom_Price_Of_Historicaldata.update_layout(template='plotly_dark')
F_gom_Price_Of_Historicaldata.show()
```

```python
GOld_Price_Of_Historicaldata[['Price','Open','High','Low']] = GOld_Price_Of_Historicaldata[['Price','Open','High','Low']].apply(pd_Price_Of_Historicaldata.to_numeric, errors ='coerce')
GOld_Price_Of_Historicaldata.info()
```

```python
GOld_Price_Of_Historicaldata.plot(subplots=True)
pm_Price_Of_Historicaldata.show()
```

```python
GOld_Price_Of_Historicaldata= GOld_Price_Of_Historicaldata[['Date',	'Price']]
print(GOld_Price_Of_Historicaldata)

GOld_Price_Of_Historicaldata.to_csv('Gold_final_data.csv', index=False)
```
